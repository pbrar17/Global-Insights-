```python
import pandas as pd

dataSetA = pd.read_csv('world.csv')
dataSetB = pd.read_csv('vacs.csv')
dataSetC = pd.read_csv('earthquake_1995-2023.csv')

# selected_columns = df.loc[:, ['Column1', 'Column2', 'Column3']]

dataSetCfiltered = dataSetC.loc[:, ['magnitude', 'date_time', 'country']]
dataSetCfiltered_by_country = dataSetCfiltered.groupby(['country'])['magnitude'].max()
dataSetA.columns = dataSetA.columns.str.upper()
dataSetB.columns = dataSetB.columns.str.upper()

# Merge the DataFrames based on the 'Country' column

merged_data = pd.merge(dataSetA, dataSetB, on='COUNTRY')
merged_data2 = pd.merge(dataSetCfiltered, dataSetCfiltered_by_country, on='country')
merged_data3 = merged_data2.drop_duplicates(subset = ['country'])
dataSetCfiltered_by_country.head()
merged_data4 = merged_data3.drop('magnitude_x', axis=1)
merged_data4.columns = merged_data4.columns.str.upper()

merge_data5 = pd.merge(dataSetA, merged_data4, on='COUNTRY')
merge_data5.head()
```

```python
m5_filtered = merge_data5.drop(['LONGITUDE', 'LATITUDE', 'CURRENCY-CODE','ABBREVIATION', 'LARGEST CITY','CALLING CODE','CAPITAL/MAJOR CITY','GROSS TERTIARY EDUCATION ENROLLMENT (%)'], axis = 1)
```

```python
m5_filtered = m5_filtered.dropna()
m5_filtered.head()
```

```python
whr = pd.read_csv("WHR2023.csv")
whr.head()
```

```python
labels = ['Very Low', 'Low', 'Moderate', 'High', 'Very High']
levels = [1,2,3,4,5]
whr['Level Of Corruption'] = pd.cut(whr['Perceptions of corruption'], bins=5, labels = labels)
whr['Level Of Social Support in Society'] = pd.cut(whr['Social support'], bins=5, labels = labels)
whr['Generosity'] = pd.cut(whr['Generosity'], bins=5, labels = levels)
whr['Freedom to make life choices'] = pd.cut(whr['Freedom to make life choices'], bins=5, labels = levels)
whr['Ladder score'] = pd.cut(whr['Ladder score'], bins=5, labels = levels)
```

```python
whr['COUNTRY'] = whr['Country name']
whr_filtered = whr[['COUNTRY', 'Generosity','Ladder score', 'Freedom to make life choices', 'Level Of Social Support in Society', 'Level Of Corruption']] 
whr_filtered.head()
```

```python

data = pd.merge(data, whr_filtered, on='COUNTRY')
```
